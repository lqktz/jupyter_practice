```python
import matplotlib.pyplot as plt
import numpy as np

# 简单的绘图
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x)) # 如果没有第一个参数 x，图形的 x 坐标默认为数组的索引
plt.show() # 显示图形
```

```python
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x),
        x,np.cos(x)) # 如果没有第一个参数 x，图形的 x 坐标默认为数组的索引
plt.show() # 显示图形
```

```python
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x),'r-o',
        x,np.cos(x),'y--') # 如果没有第一个参数 x，图形的 x 坐标默认为数组的索引
plt.show() # 显示图形
```

```python
# 使用子图
x = np.linspace(0, 2 * np.pi, 50)
#活跃区域代表当前子图所在绘图区域，绘图区域是按从左至右，从上至下的顺序编号
plt.subplot(2, 1, 1) # （行，列，活跃区）
#linewidth 线的宽度
#color='black' : 标记颜色    'g':标记颜色    'g--' :  标记颜色和线的类型
plt.plot(x, np.sin(x),color='black',linewidth=10)
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), 'g--')
plt.show()
```

```python
x = np.linspace(0,1,50)
y1 = x**2
# 打印散点图
plt.scatter(x,y1)
plt.show()
```

```python
# 彩色映射散点图
x = np.random.rand(1000)
y = np.random.rand(1000)
size = np.random.rand(1000) * 50
colour = np.random.rand(1000)
plt.scatter(x, y, size, colour)
plt.colorbar()
plt.show()
```

```python
# 添加标题，坐标轴标记和图例
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x), 'r-x', label='Sin(x)')
plt.plot(x, np.cos(x), 'g-^', label='Cos(x)')
plt.legend() # 展示图例
plt.xlabel('Rads') # 给 x 轴添加标签
plt.ylabel('Amplitude') # 给 y 轴添加标签
plt.title('Sin and Cos Waves') # 添加图形标题
plt.show()
```
